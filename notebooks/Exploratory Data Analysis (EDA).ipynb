```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors

fig, ax = plt.subplots(figsize=(6, 4), dpi=300)

# other cmap options: 'coolwarm', 'mako', 'viridis', 'rocket', 'flare', 
# 'crest', 'magma', 'cubehelix', 'Blues', 'Greens', 'Grays' etc
my_cmap = 'viridis'

scatter = sns.scatterplot(x='Learning_Rate', 
                          y='F1-score accuracy', 
                          data=my_df, 
                          hue='Run', 
                          palette=my_cmap, 
                          s=100, 
                          alpha=0.9, 
                          ax=ax)

scatter.legend_.remove()
ax.set_ylabel('R2: F1-score ACC')
ax.set_xlabel('A: Learning Rate (-)')

norm = mcolors.Normalize(vmin=my_df['Run'].min(), vmax=my_df['Run'].max())
cmap = plt.get_cmap(my_cmap)

cbar_ax = fig.add_axes([0.92, 0.15, 0.03, 0.7])
cbar = plt.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), cax=cbar_ax)
cbar.ax.yaxis.set_ticks_position('right')
cbar.ax.set_ylabel("Run", rotation=270, labelpad=15)


plt.savefig("Figure 2", dpi=300)

plt.show()
```

```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from scipy.stats import linregress

fig, ax = plt.subplots(figsize=(6, 4), dpi=300)

my_cmap = 'viridis'

# Scatter plot
scatter = sns.scatterplot(x='val_acc', 
                          y='F1-score accuracy', 
                          data=my_df, 
                          hue='Run', 
                          palette=my_cmap, 
                          s=100, 
                          alpha=0.9, 
                          ax=ax)

scatter.legend_.remove()
ax.set_ylabel('R2: F1-score ACC')
ax.set_xlabel('Validation Accuracy')

# adding a regression line:
x_reg = my_df['val_acc']
y_reg = my_df['F1-score accuracy']
slope, intercept, r_value, _, _ = linregress(x_reg, y_reg)
sns.regplot(x=x_reg, y=y_reg, scatter=False, color='gray', ax=ax, line_kws={"linestyle": "--"})
eq_text = f"y = {slope:.2f}x + {intercept:.2f}\n$R^2$ = {r_value**2:.2f}"
ax.text(0.41, 0.3, eq_text, transform=ax.transAxes, fontsize=9, bbox=dict(facecolor='white', alpha=0.0))

norm = mcolors.Normalize(vmin=my_df['Run'].min(), vmax=my_df['Run'].max())
cmap = plt.get_cmap(my_cmap)
cbar_ax = fig.add_axes([0.92, 0.15, 0.03, 0.7])
cbar = plt.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), cax=cbar_ax)
cbar.ax.yaxis.set_ticks_position('right')
cbar.ax.set_ylabel("Run", rotation=270, labelpad=15)

plt.savefig("Figure 4", dpi=300)
plt.show()
```
